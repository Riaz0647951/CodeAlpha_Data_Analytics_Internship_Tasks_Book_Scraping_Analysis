**CodeAlpha Data Analytics Internship TASK 3: Data Visualization**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for better looking plots
sns.set_style("whitegrid")

# Load the cleaned data from Task 2 (assuming you kept the same DataFrame name 'df')
try:
    df = pd.read_csv('books_data.csv')
    print("Data loaded for Visualization.")
except FileNotFoundError:
    print("Error: 'books_data.csv' not found. Please upload it again or check the filename.")
    # Stop execution if data is not available
    exit()
```

**Visualization 1: Distribution of Book Ratings**

```python
# 1. Prepare data: Convert the 'Rating_Stars' column to a string/category for better labeling
df['Rating_Label'] = df['Rating_Stars'].astype(str) + ' Star'

# 2. Create the plot
plt.figure(figsize=(8, 5))
sns.countplot(x='Rating_Label', data=df,
              order=sorted(df['Rating_Label'].unique()), # Sort by rating number
              palette='viridis')

# 3. Add Titles and Labels for clarity
plt.title('Distribution of Book Ratings Across All Pages', fontsize=16)
plt.xlabel('Star Rating', fontsize=12)
plt.ylabel('Number of Books', fontsize=12)

# 4. Display the chart
plt.show()
```

**Visualization 2: Distribution of Book Prices**

```python
# Create a figure with two subplots side-by-side
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# --- Subplot 1: Histogram (to show frequency distribution) ---
sns.histplot(df['Price'], bins=30, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Frequency Distribution of Book Prices', fontsize=14)
axes[0].set_xlabel('Price (£)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)

# --- Subplot 2: Box Plot (to identify outliers and central tendency) ---
sns.boxplot(y=df['Price'], ax=axes[1], color='lightcoral')
axes[1].set_title('Box Plot of Book Prices (Outlier Detection)', fontsize=14)
axes[1].set_ylabel('Price (£)', fontsize=12)

# 3. Adjust layout and display the charts
plt.tight_layout()
plt.show()
```

**Visualization 3: Average Price by Star Rating**

```python
# 1. Create a figure
plt.figure(figsize=(10, 6))

# 2. Create the plot using seaborn's barplot, which calculates the mean and error bars automatically
# x='Rating_Stars' is the independent variable (category), y='Price' is the dependent variable (value)
sns.barplot(x='Rating_Label', y='Price', data=df,
            order=sorted(df['Rating_Label'].unique()),
            palette='coolwarm')

# 3. Add Titles and Labels
plt.title('Average Book Price by Star Rating', fontsize=16)
plt.xlabel('Star Rating', fontsize=12)
plt.ylabel('Average Price (£)', fontsize=12)
plt.yticks(rotation=0)

# 4. Display the chart
plt.show()
```
